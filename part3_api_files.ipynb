# ***Task 2: API Integration***

```python
!pip install requests

import requests

#Step1:- Fetch 20 Products
url = "https://dummyjson.com/products?limit=20"

response = requests.get(url)
data = response.json()
products = data["products"]

#Printing Table Header
print(f"{'ID':<4} | {'Title' :<30} |{'Category':<15} | {'Price': <8} | Rating")
print("-" * 75)

for product in products:
  print(f"{product['id']:<4} | {product['title']:<30} | {product['category']:<15} | ${product['price']:<8} | {product['rating']}")


#Step2:- Filter Rating >= 4.5 and sort by price
filtered = []
for product in products:
  if product["rating"] >= 4.5:
    filtered.append(product)

#Sort price by desecending
sorted_products = sorted(filtered, key=lambda x: x["price"], reverse=True)
print("\n Filtered (rating >= 4.5) & Sorted by Price:")
print("-" * 75)

for product in sorted_products:
  print(f"{product['id']:<4} | {product['title']:<30} | {product['category']:<15} | ${product['price']:<8}")

#Step3:- Fetching Laptops Category
laptop_url = "https://dummyjson.com/products/category/laptops"
laptop_response = requests.get(laptop_url)
laptop_data = laptop_response.json()
laptops = laptop_data["products"]

print("\nLaptops:")
print("-" * 40)
for laptop in laptops:
  print(f"{laptop['title']} -${laptop['price']}")

#Step4:- POSTING Request
post_url = "https://dummyjson.com/products/add"

new_product = {
    "title":"My Custom Product",
    "price": 999,
    "category": "electronics",
    "description": "A product I created via API"
}

post_response = requests.post(post_url, json=new_product)
print("\nPost Response:")
print(post_response.json())

```

# ***Task3: Exceptional Handling***

```python
import requests

#PartA:- Safe Divide Function
def safe_divide(a, b):
  try:
    return a / b
  except ZeroDivisionError:
    return "Error : Cannot divide by zero"
  except TypeError:
    return "Error : Invaild input types"

#Testing
print("Part A - Safe Divide:")
print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))


#PartB:- Safe File Reader
def read_file_safe(filename):
  try:
    file = open(filename,"r",encoding="utf-8")
    content = file.read()
    file.close()
    return content
  except FileNotFoundError:
    print(f"Error: File '{filename}' not found.")
  finally:
    print("file operation attempt complete.")

#Testing
print("\nPart B - Safe File Reader:")
print("Reading existing file:")
print(read_file_safe("python_notes.txt"))

print("Reading non-existing file:")
print(read_file_safe("ghost_file.txt"))


#PartC: API Calls with Exception Handling
def fetch_products():
  try:
    response = requests.get("https://dummyjson.com/products?limit=20",timeout=5)
    return response.json()["products"]
  except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internets.")
  except requests.exceptions.Timeout:
    print("Request timed out. Please try again later.")
  except Exception as e:
    print(f"An error occurred: {e}")

def fetch_laptops():
  try:
    response = requests.get("https://dummyjson.com/products/category/laptops", timeout=5)
    return response.json()["products"]
  except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet connection.")
  except requests.exceptions.Timeout:
    print("Request timed out. Please try again later.")
  except Exception as e:
    print(f"An error occurred: {e}")

print("\n PartC: API Calls with Exception Handling:")
products = fetch_products()
if products:
  print(f"Fetched {len(products)} products successfully.")

#PartD: Input Vaildation Loop
print("\n PartD - Product Lookup:")
while True:
  user_input = input("Enter product ID (1-100) or 'q' to quit: ")
  if user_input.lower() == 'q':
    print("GoodBye!")
    break
  try:
    product_id = int(user_input)
  except ValueError:
    print("Warning: Enter number only!")
    continue

  if product_id < 1 or product_id > 100:
    print("Warning: Enter number between 1")
```

# ***Task4: Logging to File***


```python
from datetime import datetime
import requests

def log_error(function_name,error_type,message):
  timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
  log_entry = f"{timestamp} - Function: {function_name} - Error Type: {error_type} - Message: {message}\n"
  print(f"Error logged: {log_entry}")
  file = open("error_log.txt", "w", encoding="utf-8")
  file.write("")
  file.close()



  #=======Trigger 1: ConnectionError========
def fetch_bad_url():
    try:
      response = requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
    except requests.exceptions.ConnectionError:
      log_error("fetch_bad_url", "ConnectionError", "No connection could be made")
    except requests.exceptions.Timeout:
       log_error("fetch_bad_url", "Timeout", "Request timed out")
print("Triggering ConnectionErro........")
fetch_bad_url()

  #========Trigger 2: HTTp 404 ERROR========
def lookup_products(product_id):
    try:
      response = requests.get(f"https://dummyjson.com/products/{product_id}", timeout=5)
      if response.status_code == 200:
        product = response.json()
        print(f"Found: {product['title']} - ${product['price']}")
      elif response.status_code == 404:
        print(f"Product {product_id} not found.")
        log_error("lookup_product", "HTTPError", f"404 Not Found for product ID {product_id}")
    except requests.exceptions.ConnectionError:
      log_error("lookup_product", "ConnectionError", "No connection could be made")
    except requests.exceptions.Timeout:
      log_error("lookup_product", "Timeout", "Request timed out")
print("\n Triggering 404 Error......")
lookup_products(999)

#========Prininting Error Log File=========
print("\n--- Error Log Contents ---")
file = open("error_log.txt", "r", encoding="utf-8")
content = file.read()
file.close()
print(content)
```
